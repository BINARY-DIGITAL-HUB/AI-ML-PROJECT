```python
# capture image from video 
cam = cv2.VideoCapture(0)

print(cam.get(0))

while(cam.isOpened()): 
    
    # read frame by frame
    ret, frame = cam.read()

    # print('Width :' , cam.get(3), '  hieght : ', cam.get(4))
    
    # do any operation with the frame 
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)
    
    cv2.imshow('vid', img)

    # breaking the loop condition
    key = cv2.waitKey(1)
    if key & 0xFF == ord('e'): 
        break

# release resource 
cam.release()
cv2.destroyAllWindows()
```

```python
# face image detection
import cv2
import numpy as np 
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
img = cv2.imread('img.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3)
for (x,y,w,h) in faces: 
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# initializaing face detection model 
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')


# capture image from video 
cam = cv2.VideoCapture(0)



while(cam.isOpened()): 

   
    # read frame by frame
    ret, frame = cam.read()

    
    # do any operation with the frame 
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)

    # detecting face location from the vid cam
    faces = face_cascade.detectMultiScale(img, 1.3, 2)

    # print('Width :' , cam.get(3), '  hieght : ', cam.get(4))
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
    
    
    
    cv2.imshow('vid', img)

    # breaking the loop condition
    key = cv2.waitKey(1)
    if key & 0xFF == ord('e'): 
        break

# release resource 
cam.release()
cv2.destroyAllWindows()
```
