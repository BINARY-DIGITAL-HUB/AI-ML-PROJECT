```python
import random


rand_val = list(range(30 , 100))

com_course = ['CPT111', 'CPT112' , 'CPT113' , 'CPT114' , 'CPT211', 'CPT212' , 'CPT213' , 'CPT214' , 'CPT311', 'CPT312' , 'CPT313' , 'CPT314', 'CPT411', 'CPT412' , 'CPT413' , 'CPT414']
css_course = ['CSS111', 'CSS112' , 'CSS113' , 'CSS114' , 'CSS211', 'CSS212' , 'CSS213' , 'CSS214' , 'CSS311', 'CSS312' , 'CSS313' , 'CSS314', 'CSSM411', 'CSS412' , 'CSS413' , 'CSS414']
imt_course = ['IMT111', 'IMT112' , 'IMT113' , 'IMT114' , 'IMT211', 'IMT212' , 'IMT213' , 'IMT214' , 'IMT311', 'IMT312' , 'IMT313' , 'IMT314', 'IMT411', 'IMT412' , 'IMT413' , 'IMT414']





def add_column(df, name, val):
    r, c = df.shape
    df.insert(loc = c, column=name,
              value=val)
    return df


for index in range(len(com_course)): 
    
    valcom = [random.choice(rand_val) for val in range(len(com_data))]
    valcss =[random.choice(rand_val) for val in range(len(css_data))]
    valimt = [random.choice(rand_val) for val in range(len(imt_data))]
    
#   =============================================================================
    com = add_column(com_data, com_course[index], valcom)
    imt = add_column(imt_data, imt_course[index], valimt)
    css = add_column(css_data, css_course[index], valcss)
#   ============================================================================


    
```

```python
def cover_score_to_grade(score):
    
    if score >= 0 and score < 40 : 
        grade = 'F'
    elif score >= 40 and score < 49 : 
        grade = 'E'
    elif score >= 50 and score < 60 :
        grade = 'C'
    elif score >= 60 and score < 75 :
        grade = 'B'
    elif score >= 75 and score <= 100 : 
        grade = 'A'
    else: 
        grade='F'
        
    return grade

grade = cover_score_to_grade(50)   
grade
```

```python
# imt[imt.columns[3:7]]
# imt[imt.columns[7:11]]
# imt[imt.columns[11:15]]
# imt[imt.columns[15:19]]

def get_level_data(dept=com , level = 100 , start=3, stop=7):
    
    data = None 
    
    if level == 100:    
        data = dept[dept.columns[4:8]]
    elif level == 200: 
         data = dept[dept.columns[8:12]]
    elif level == 300: 
        data = dept[dept.columns[12:16]]
    elif level == 400:
       data = dept[dept.columns[16:20]]
    else : 
        data = None
        
    return data

level_data = get_level_data(com , 100)[:40]


level_data
```

```python
def convert_to_grade_table(level_data):

    level_grade_list =  []
    for index in range(len(level_data)): 
        d=level_data.values[index]
        grade = [cover_score_to_grade(dm) for dm in d] 
        level_grade_list.append(grade)
    #     print(grade)

    level_grade_list = np.array(level_grade_list)
    level_grade_list

    df = pd.DataFrame(level_grade_list , columns=level_data.columns)
    return df 


data = convert_to_grade_table(level_data)


data
```

```python
def get_course_score_statistic(data, course): 
    
    grade_list = ['A', 'B', 'C' , 'D', 'E', 'F']
    diction_data = {}
    diction_percentatage = {}
    total_count = []  

    for index in range(len(grade_list)):
                       
        count = data[course][data[course]== grade_list[index]].count()
        diction_data[grade_list[index]] = count
        total_count.append(count)

    
    total_count= np.array(total_count)
#     print(total_count.sum())
    percent_distribution = [(cnt/total_count.sum()) * 100 for cnt in total_count]
    
    
    return diction_data , percent_distribution
                       
                       
gr , pr = get_course_score_statistic(data, 'COM314')                       
```
