```python
# define gradient decent function 
def gradient_decent(x, y, lr=0.08, iter=1000): 
    m_current , b_current = 0 , 0
    for i in range(iter):
        y_pred = m_current*x + b_current
        n = len(x)
        # calculate change in m and b 
        m = -(2/n) * sum(x * (y - y_pred))
        b = -(2/n) * sum(y - y_pred)

        # cost (mse)
        mse = (1/n) * np.square(sum(y-y_pred))
    
        # update m and b current
        m_current = m_current - lr * m
        b_current = b_current - lr * b 
    
        # print training 
        print(f'====== M: {m_current} ===== B: {b_current} === Error: {mse} ===== Iteration : {i}')

    
gradient_decent(X, y)
    
```
