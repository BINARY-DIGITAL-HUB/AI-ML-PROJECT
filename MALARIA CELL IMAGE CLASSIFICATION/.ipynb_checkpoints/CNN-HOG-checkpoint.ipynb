```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import PIL
import os
import cv2
import matplotlib.pyplot as plt
import pathlib
import pandas as pd 
```

## DATA LOADING 

```python
print(os.listdir())
```

```python
cell_image_path = './/cell_images'

window_path = pathlib.Path(cell_image_path)
window_path
```

```python
# dataset_path = list(window_path.glob('*/*.png'))
# print ('Lenght of image dataset ======> ', len(dataset_path))

# accesssing individual folders (loading parasite folder)
parasitize_folder = list(window_path.glob('Parasitized/*.png'))

# loading infected maleria folder
uninfected_folder = list(window_path.glob('Uninfected/*.png'))
```

```python
# create a diction of label and images set 

cell_image_dictionary = {
    'parasitized':parasitize_folder, 
    'uniffected':uninfected_folder
}

cell_image_label = {
    'parasitized':0,
    'uniffected':1
}
```

```python
sample_image  = cv2.imread(str(cell_image_dictionary['parasitized'][0]))
print(sample_image.shape)

# array defination of image
def plot_image(image , width, hieght, title=''): 
    plt.figure(figsize=(width, hieght))
    plt.imshow(image)
    plt.title(title)
    plt.show()
    
plot_image(sample_image, 4, 4, 'infected cell')
```

```python
unfected = str(cell_image_dictionary['uniffected'][0])
uninfected_sample = cv2.imread(unfected)
plot_image(uninfected_sample, 4,4, 'Uninfected sample')
```

```python
# USING THE DICTIONARY CREATED ABOVE TO LOAD THE ENTIRE IMAGE SET 

X = [] # actual images are stored in this variable....
y = [] # labels are stored in this variable...........

for title, paths in cell_image_dictionary.items():
    
    print('Loading Path =============> : ' , title)
    for path in paths:
        image = cv2.imread(str(path))
        X.append(image)
        y.append(cell_image_label[title])

print(f'{len(X)} Images Loaded completely .....................')
```

## EXPLORING DATASET

```python
# print(_X.shape)
X_padded = []
for x in _X: 
    X_padded.append(cv2.resize(x, (64, 64)))
```

```python
[x.shape for x in X_padded[:5]]
```

```python
# SCALLING 
padded_X = np.array(padded_X) / 255
```

```python
from sklearn import  model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(padded_X, _y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
prediction_class = 1

cnn_model = keras.Sequential([
#     keras.layers.Conv2D(filters=64 , kernel_size=(3,3) , padding='same', activation='relu', input_shape=(64,64,3)),
    keras.layers.Conv2D(filters=32 , kernel_size=(3,3) , padding='same', activation='relu', input_shape=(64,64,3)),
    keras.layers.MaxPooling2D(),
    
    keras.layers.Conv2D(filters=64 , kernel_size=(3,3) , padding='same', activation='relu'),
    keras.layers.MaxPooling2D(), 
    
    keras.layers.Flatten(),
#     keras.layers.Dropout(.50),
    #     keras.layers.Dropout(.50),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(prediction_class, activation='sigmoid')
    
])



cnn_model.compile(
    loss = "binary_crossentropy",
    optimizer='adam',
    metrics=['accuracy']
)
```
