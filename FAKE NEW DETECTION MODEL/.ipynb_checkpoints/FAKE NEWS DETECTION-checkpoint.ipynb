# DATA CLEANING

```python
# cleaning the text corpus
import string 
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
st = stopwords.words('english')

# remove new line character and special symbols
def clean_text(doc):
    doc = ''.join([d.replace('\n', '@').replace('—', '@') for d in doc])
    doc = ''.join([d.lower() for d in doc if d not in string.punctuation])
    doc = doc.split(' ')
   

#     token = word_tokenize(doc)
#     doc = [stemmer.stem(d) for d in token if not d in string.punctuation] 
#     doc = [d for d in doc if not d in st]
    
    return doc

news['clean_text'] = news['text'].apply(lambda x: clean_text(x))
news.head()
```

```python
clean_news = news[['label', 'clean_text']]
clean_news.head()
```

## DATA TRANSFORMATION

```python
import warnings
warnings.filterwarnings('ignore')

clean_news.label.replace({'FAKE':0, 'REAL':1}, inplace=True)
clean_news.head()
```

## DATA SPLITTING

```python
from sklearn import model_selection

xtrain, xtest, ytrain, ytest = model_selection.train_test_split(clean_news.clean_text, clean_news.label, test_size=.20)
print(f'train x : {len(xtrain)}' )
print(f'train y : {len(ytrain)}' )
print(f'test  x : {len(xtest)}' )
print(f'test  y : {len(ytest)}' )
```

```python
# VECTORIZING THE CLEAN_TEXT COLUMN
# deep leanring transformation 
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 10000

tokenizer = Tokenizer(num_words=10000, oov_token='<00V>')
tokenizer.fit_on_texts(xtrain)
print(len(tokenizer.word_index))

train_sequence = tokenizer.texts_to_sequences(xtrain)
test_sequence = tokenizer.texts_to_sequences(xtest)

# padding the sequences 
max_val = 5000
train_pad = pad_sequences(train_sequence, maxlen=max_val)
text_pad = pad_sequences(test_sequence, maxlen=max_val)
train_pad.shape
```

```python

# vocab_size = 10000
embedding_dim = 100
pad_length =5000

en_lstm_model = tf.keras.Sequential([
    
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=pad_length),
#   tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, dropout=0.2, recurrent_dropout=0.5)),
    tf.keras.layers.LSTM(100, dropout=0.5),
    
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(5, activation='sigmoid')
    
])


en_lstm_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'] )
en_lstm_model.summary()
```

```python
en_lstm_model.fit(train_pad, ytrain , epochs=3, validation_data=(text_pad, ytest), verbose=1)
```
