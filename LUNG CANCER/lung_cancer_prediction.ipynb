### Model Traning

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics

# creating helper function 
def model_prediction_and_report(model, xtrain, ytrain, xtest, ytest):
    data_holder = {'train_score':None, 'test_score':None, 'acc':None, 'pre':None, 'rec':None, 'report':None, 'matrix':None, 'prediction':None}
    model.fit(xtrain, ytrain)
    train_score= model.score(xtrain, ytrain)
    test_score = model.score(xtest, ytest)
    prediction = model.predict(xtest)
    acc = metrics.accuracy_score(ytest , prediction)
    pre = metrics.precision_score(ytest , prediction, average='weighted')
    rec = metrics.recall_score(ytest , prediction, average='weighted')
    report = classification_report(ytest, prediction)
    matrix = confusion_matrix(ytest, prediction)

    # data project 
    data_holder['train_score'] = train_score
    data_holder['test_score'] = test_score
    data_holder['acc'] = acc
    data_holder['pre'] = pre
    data_holder['rec'] = rec
    data_holder['report'] = report
    data_holder['matrix'] = matrix
    data_holder['prediction'] = prediction

    return data_holder
    
```

```python
# Support Vector Machine... 
svm = SVC()
svm_report = model_prediction_and_report(svm, X_train, y_train, X_test, y_test)
```
