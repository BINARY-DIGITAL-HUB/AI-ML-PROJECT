# Data Transformation and Preparation

```python
dataset = data.drop(columns=['AQI Value', 'CO AQI Category', 'Ozone AQI Category', 'NO2 AQI Category', 'PM2.5 AQI Category'])
dataset
```

```python
dataset.info()
```

```python
dataset['Country']  =  dataset.Country.astype('category')
dataset['City']  =  dataset.City.astype('category')
dataset.info()
```

```python
dataset['City_code'] = dataset.City.cat.codes
dataset['Country_code'] =dataset.Country.cat.codes
```

```python
dataset
```

```python
niger_data = dataset[dataset['Country'] == 'Nigeria']
nig_data = niger_data.drop(columns=['Country', 'City'])
nig_data['AQI Category'] = nig_data['AQI Category'].astype('category').cat.codes
nig_data
```

## extracting dependent and independent variable

```python
y = dataset['AQI Category']
X = dataset.drop(columns=['AQI Category', 'Country' , 'City'])
```

```python
# # Scalling the independent variable
# mm_scaler = MinMaxScaler()
# mm_scaler.fit_transform(X[X.columns[1:]])

dataset['AQI Category'] = dataset['AQI Category'].astype('category')
dataset.info()
```

```python
dataset['AQI Category'] = dataset['AQI Category'].cat.codes
```

```python
import seaborn as sn 

dt_cor = dataset.drop(columns=['Country' , 'City'])
dt_cor.corr()
sn.heatmap(dt_cor.corr(), annot=True)
plt.title('Data Correlation')
plt.show()
```

```python
labels = ['Good', 'Hazardous', 'Moderate', 'Unhealthy', 'Unhealthy for Sensitive Groups', 'Very Unhealthy']
```

#### Label Encoding (Converting label into numeric)

```python
# transforming the dependent variable (y) into numeric......
lb_encode = LabelEncoder()
y_code = lb_encode.fit_transform(y)
y_code = dataset['AQI Category']
y_code
```

```python
X = X.drop(columns=["City_code", "Country_code"])
```

```python
X
```

#### data spliting

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X ,y_code, test_size=0.2)

print('Xtrain size : {}'.format(len(X_train)))
print('Ytrain size : {}'.format(len(y_train)))
print('Xtest size :  {}'.format(len(X_test)))
print('Ytest size :  {}'.format(len(y_test)))
```

### MACHINE LEARNING MODEL

```python
from sklearn import metrics
import seaborn as sn
```

#### Support Vector Machine

```python
svm_model = svm.SVC(kernel='linear')
svm_model = svm_model.fit(X_train, y_train)
```

```python
svm_score = svm_model.score(X_test, y_test)
svm_score
```

```python
sv_predictions = svm_model.predict(X_test)
report = metrics.classification_report(y_test, sv_predictions)
confustion_mx = metrics.confusion_matrix(y_test, sv_predictions)
# sv_predictions
print(report)
```

```python
sn.heatmap(confustion_mx, fmt='d' , annot=True, xticklabels=labels, yticklabels=labels)
```

## ANN Model

```python
import tensorflow as tf

data_size = len(X_test.columns)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(data_size, )), 
    tf.keras.layers.Dense(500 , activation='relu'), 
    tf.keras.layers.Dense(200 , activation='relu'),
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(6, activation='softmax')
    ])

model.compile(
        loss='sparse_categorical_crossentropy', 
        optimizer='Adam',
        metrics=['Accuracy']
    )

model.summary()
```

```python
training_history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10)
```

```python
training_history.history.keys()
```

```python
ann_training_acc = training_history.history['Accuracy']
ann_testing_accc = training_history.history['val_Accuracy']
ann_training_loss = training_history.history['loss']
ann_testing_loss = training_history.history['val_loss']

plt.plot(ann_training_acc, label='training_acc')
plt.plot(ann_testing_accc, label='testing_acc')
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.title("ANN ACCURACY TRAINING PLOT")
plt.legend()
plt.show()
```

```python
plt.plot(ann_training_loss, label='training_loss')
plt.plot(ann_testing_loss, label='testing_loss')
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.title("ANN LOSS PLOT")
plt.legend()
plt.show()
```

```python
ann_result = model.evaluate(X_test, y_test)
```

```python
predict = model.predict(X_test)
predict = [np.argmax(pre) for pre in predict]
```

```python
sn.heatmap(ann_confu, fmt='.1f' , annot=True, xticklabels=labels, yticklabels=labels)
plt.ylabel("Actual Prediction")
plt.xlabel("Predicted Value")
plt.title('ANN Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier


knn = KNeighborsClassifier()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
ann = MLPClassifier()



models = [('gb',gb), ('rf', rf), ('knn', knn), ('ann', ann) , ]

# constructing the assemble model
ensemble_model =  VotingClassifier(
estimators=models , voting='hard')

# training the ensemple model
ensemble_model.fit(X_train, y_train)
```

```python
# testing the model 
ensemble_model.score(X_test, y_test)
```

```python
ensemble_prediction = ensemble_model.predict(X_test)
```

```python
print("ENSEMBLE CLASSIFIER TRAINING REPORT")
print("==========================================")
print("ENSEMBLE MODEL Accuracy :  {}".format(metrics.accuracy_score(ensemble_prediction, y_test,)))
print("ENSEMBLE MODEL Precision : {} ".format(metrics.precision_score(ensemble_prediction, y_test, average='weighted')))
print("ENSEMBLE MODEL Recall  :   {}".format(metrics.recall_score(ensemble_prediction, y_test, average='weighted')))
print("ENSEMBLE MODEL F1-score  : {}".format(metrics.f1_score(ensemble_prediction, y_test, average='weighted')))
```
